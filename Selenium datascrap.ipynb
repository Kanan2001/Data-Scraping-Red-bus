Bus route and bus name data scraping

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException
import pandas as pd
import time

```

```python


# Initialize the Chrome WebDriver
driver = webdriver.Chrome()

# Open the RedBus KSRTC Kerala webpage
driver.get("https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile")
driver.maximize_window()  # Maximize the browser window for better visibility

# Initialize WebDriverWait with a timeout of 20 seconds
wait = WebDriverWait(driver, 20)

def kerala_linkroute(path):
    linkkerala = []  # List to store route links
    routekerala = []  # List to store route names
 
    # Iterate through the first two pages
    for i in range(1, 3):
        # Find all elements matching the given XPath for route links
        paths = driver.find_elements(By.XPATH, path)
        
        # Extract href attributes (links) from the found elements and add them to the linkkerala list
        for links in paths:
            a = links.get_attribute("href")
            linkkerala.append(a)
        
        # Extract text (route names) from the found elements and add them to the routekerala list
        for route in paths:
            routekerala.append(route.text)
        
        try:
            # Wait until the pagination table is present
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
            # Wait until the next page button is clickable
            next_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, f'div.DC_117_pageTabs:nth-child({i+1})')))
            
            # Scroll to the next button (optional but useful for visibility)
            driver.execute_script("arguments[0].scrollIntoView(true);", next_button)
                    
            time.sleep(5)  # Pause to ensure the next button is fully loaded
                    
            # Click the next page button to load the next set of results
            next_button.click()
                
        except NoSuchElementException:
            # Handle the case where there are no more pages to paginate
            print(f'No more pages to paginate at step {i}')
            break
        except TimeoutException:
            # Handle the case where the next page button was not found in time
            print(f'Next page button not found for page {i}')
            break
        except Exception as e:
            # Handle any other exceptions that occur
            print(f'An error occurred: {e}')

    # Return the collected route links and names
    return linkkerala, routekerala

# Call the function with the XPath for route links and store the results in linkkerala and routekerala
linkkerala, routekerala = kerala_linkroute('//a[@class="route"]')

# Create a DataFrame from the collected data
df_kerala = pd.DataFrame({'Routename': routekerala, 'Routelink': linkkerala})

# Define the path to save the CSV file
path = r"D:\.intel\.shared\Redbus\Bus route\df_Kerala.csv"
# Save the DataFrame to a CSV file
df_kerala.to_csv(path, index=False)

# Close the browser window
driver.quit()
```

```python


# Read the CSV file containing route links and names
df = pd.read_csv("D:/.intel/.shared/Redbus/Bus route/df_Kerala.csv")

# Initialize lists to store extracted data
Bus_names_kerala = []
Route_links = []
Routes_name = []
Bus_types_kerala = []
Price_kerala = []
Start_time_kerala = []
End_time_kerala = []
Total_Duration_kerala = []
Ratings_kerala = []
Seats_kerala = []
data = []

# Initialize the Chrome driver
driver_kerala = webdriver.Chrome()
driver_kerala.maximize_window()

def scroll_and_load(driver):
    """
    Scroll down the page to load more content.
    """
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)  # Wait for the new content to load
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break  # Exit loop if no more new content is loaded
        last_height = new_height

# Iterate through each row in the DataFrame
for i, r in df.iterrows():
    link = r["Routelink"]  # Extract route link
    routes = r["Routename"]  # Extract route name

    # Open the route link in the browser
    driver_kerala.get(link)
    time.sleep(5)  # Wait for the page to load

    # Click a specific button to view more details, if it exists
    try:
        clicks = driver_kerala.find_element(By.XPATH, "//div[@class='button']")
        clicks.click()
        time.sleep(3)  # Wait for additional details to load
    except Exception as e:
        print(f"Exception encountered while clicking button: {e}")
        continue  # Skip to the next route if an error occurs

    # Scroll down the page to load more content
    scroll_and_load(driver_kerala)

    # Extract bus details using XPath
    try:
        bus_name_elements = driver_kerala.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
        bus_type_elements = driver_kerala.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
        start_time_elements = driver_kerala.find_elements(By.XPATH, "//div[@class='dp-time f-19 d-color f-bold']")
        end_time_elements = driver_kerala.find_elements(By.XPATH, "//div[@class='bp-time f-19 d-color disp-Inline']")
        total_duration_elements = driver_kerala.find_elements(By.XPATH, "//div[@class='dur l-color lh-24']")
        price_elements = driver_kerala.find_elements(By.XPATH, "//div[@class='fare d-block']")
        seats_elements = driver_kerala.find_elements(By.XPATH, "//div[@class='column-eight w-15 fl']")
        ratings_elements = driver_kerala.find_elements(By.XPATH, "//div[@class='clearfix row-one']//div[@class='column-six p-right-10 w-10 fl']")

        # Append extracted details to the lists
        for j in range(len(bus_name_elements)):
            Bus_names_kerala.append(bus_name_elements[j].text if j < len(bus_name_elements) else "")
            Route_links.append(link)
            Routes_name.append(routes)
            Bus_types_kerala.append(bus_type_elements[j].text if j < len(bus_type_elements) else "")
            Price_kerala.append(price_elements[j].text if j < len(price_elements) else "")
            Start_time_kerala.append(start_time_elements[j].text if j < len(start_time_elements) else "")
            End_time_kerala.append(end_time_elements[j].text if j < len(end_time_elements) else "")
            Total_Duration_kerala.append(total_duration_elements[j].text if j < len(total_duration_elements) else "")
            Ratings_kerala.append(ratings_elements[j].text if j < len(ratings_elements) else "")
            Seats_kerala.append(seats_elements[j].text if j < len(seats_elements) else "")

    except Exception as e:
        print(f"Exception encountered while extracting details: {e}")

# Ensure all lists are of the same length before creating the DataFrame
max_len = min(len(Bus_names_kerala), len(Route_links), len(Routes_name), len(Bus_types_kerala), len(Price_kerala), len(Start_time_kerala), len(End_time_kerala), len(Total_Duration_kerala), len(Ratings_kerala), len(Seats_kerala))

# Truncate lists to the length of the smallest list to avoid mismatched data
Bus_names_kerala = Bus_names_kerala[:max_len]
Route_links = Route_links[:max_len]
Routes_name = Routes_name[:max_len]
Bus_types_kerala = Bus_types_kerala[:max_len]
Price_kerala = Price_kerala[:max_len]
Start_time_kerala = Start_time_kerala[:max_len]
End_time_kerala = End_time_kerala[:max_len]
Total_Duration_kerala = Total_Duration_kerala[:max_len]
Ratings_kerala = Ratings_kerala[:max_len]
Seats_kerala = Seats_kerala[:max_len]

# Create a list of dictionaries with the extracted data
for i in range(max_len):
    seats_split = Seats_kerala[i].split()
    seats_value = seats_split[0] if seats_split else "No data"  # Extract seats value, handle cases with no data
    data.append({
        'Bus_name': Bus_names_kerala[i],
        'Route_link': Route_links[i],
        'Routes_name': Routes_name[i],
        'Bus_types': Bus_types_kerala[i],
        'Price': Price_kerala[i],
        'Start_time': Start_time_kerala[i],
        'End_time': End_time_kerala[i],
        'Total_Duration': Total_Duration_kerala[i],
        'Rating': Ratings_kerala[i],
        'Seats': seats_value
    })

# Convert the list of dictionaries to a DataFrame
df_buses_1kl = pd.DataFrame(data)

# Define the file path where the CSV file will be saved
csv_file_path = 'D:/.intel/.shared//Redbus/Bus Details/df_buses_1kl.csv'

# Save the DataFrame to a CSV file
df_buses_1kl.to_csv(csv_file_path, index=False)

# Close the browser
driver_kerala.quit()
```
